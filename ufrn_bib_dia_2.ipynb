```python
import pandas as pd
```

# Tratamentos dia 2:
  Criar lista do CDU
  Incluir coluna para nomes do CDU
  Importar dados do CDU para preencher a nova coluna como as decrições correpondentes

```python
# Criar um lista contendo as seguintes informações: CDU - Classificação Decimal Universal. Esta classificação é decimal, pois varia de acordo com a classe de cada assunto:
# 000 a 099: Generalidades. Ciência e conhecimento.
# 100 a 199: Filosofia e psicologia.
# 200 a 299: Religião.
# 300 a 399: Ciências sociais.
# 400 a 499: Classe vaga. Provisoriamente não ocupada.
# 500 a 599: Matemática e ciências naturais.
# 600 a 699: Ciências aplicadas.
# 700 a 799: Belas artes.
# 800 a 899: Linguagem. Língua. Linguística.
# 900 a 999: Geografia. Biografia. História."

cdu_lista = [
    {'codigo': '000 a 099', 'descricao': 'Generalidades. Ciência e conhecimento.'},
    {'codigo': '100 a 199', 'descricao': 'Filosofia e psicologia.'},
    {'codigo': '200 a 299', 'descricao': 'Religião.'},
    {'codigo': '300 a 399', 'descricao': 'Ciências sociais.'},
    {'codigo': '400 a 499', 'descricao': 'Classe vaga. Provisoriamente não ocupada.'},
    {'codigo': '500 a 599', 'descricao': 'Matemática e ciências naturais.'},
    {'codigo': '600 a 699', 'descricao': 'Ciências aplicadas.'},
    {'codigo': '700 a 799', 'descricao': 'Belas artes.'},
    {'codigo': '800 a 899', 'descricao': 'Linguagem. Língua. Linguística.'},
    {'codigo': '900 a 999', 'descricao': 'Geografia. Biografia. História.'}
]
```

```python
cdu_lista
```

```python
dados_emprestimos = pd.read_csv('/content/sample_data/df_concat_cleaned.csv')
dados_emprestimos
```

```python
dados_emprestimos_limpo = dados_emprestimos.drop_duplicates()
dados_emprestimos_limpo
```

```python
# Mudando o tipo de dados da coluna localizacao para int64 sem cas decimal

dados_emprestimos_limpo['localizacao'] = pd.to_numeric(dados_emprestimos_limpo['localizacao'], errors='coerce').fillna(0).astype('int64')
```

```python
dados_emprestimos_limpo
```

```python
# Preenchendo a coluna classe_cdu com o campo descricao, correpondentes ao arrango de codigo  conforme os limites nesse campo

def preencher_classe_cdu(row):
  localizacao = row['localizacao']
  for item in cdu_lista:
    codigo_inicio, codigo_fim = item['codigo'].split(' a ')
    codigo_inicio = int(codigo_inicio)
    codigo_fim = int(codigo_fim)

    if codigo_inicio <= localizacao <= codigo_fim:
      return item['descricao']
  return None  # Retorna None se não encontrar correspondência

# Aplica a função para preencher a coluna 'classe_cdu'
dados_emprestimos_limpo['classe_cdu'] = dados_emprestimos_limpo.apply(preencher_classe_cdu, axis=1)
```

```python
dados_emprestimos_limpo
```

```python
# Excluindo a coluna registro_sistema do dataframe dados_empresimos_limpo

dados_emprestimos_limpo = dados_emprestimos_limpo.drop('registro_sistema', axis=1)
```

```python
dados_emprestimos_limpo
```

```python
# Mudando o tipo de dados da coluna matricula_ou_siape para string sem as casas decimais

dados_emprestimos_limpo['matricula_ou_siape'] = dados_emprestimos_limpo['matricula_ou_siape'].astype(str).str.split('.').str[0]
```
