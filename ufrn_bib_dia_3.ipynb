```python
# Alterando colunas data_emprestimos e data_devolucao para data em formato do Brasil somente ano

import pandas as pd
dados_emprestimos = pd.DataFrame()
dados_emprestimos = pd.read_csv('/content/sample_data/dados_emprestimos.csv')
dados_emprestimos.value_counts()
dados_emprestimos


# Converter as colunas para datetime
dados_emprestimos['data_emprestimo'] = pd.to_datetime(dados_emprestimos['data_emprestimo'])
dados_emprestimos['data_devolucao'] = pd.to_datetime(dados_emprestimos['data_devolucao'])
dados_emprestimos['data_renovacao'] = pd.to_datetime(dados_emprestimos['data_renovacao'])


# Extrair somente o ano e formatar as datas
dados_emprestimos['data_emprestimo'] = dados_emprestimos['data_emprestimo'].dt.year
dados_emprestimos['data_devolucao'] = dados_emprestimos['data_devolucao'].dt.year
dados_emprestimos['data_renovacao'] = dados_emprestimos['data_renovacao'].dt.year

dados_emprestimos.head()
```

```python
# Transformando a coluna matricula_ou_siape em formato string

# Transformar a coluna matricula_ou_siape em formato string
dados_emprestimos['matricula_ou_siape'] = dados_emprestimos['matricula_ou_siape'].astype(str)
```

```python
# Agrupando por ano e contando pela coluna id_exemplar

emprestimos_por_ano = dados_emprestimos.groupby('data_emprestimo')['id_exemplar'].count()
emprestimos_por_ano
```

```python
# prompt: Criar gráfico de linhas com eixo x=data_emprestimo e y=id_exemplar, com título 'Emprestimos de exemplares por cada ano - 2010 a 2020

import matplotlib.pyplot as plt

# Criando o gráfico de linhas
plt.plot(emprestimos_por_ano.index, emprestimos_por_ano.values)

# Configurando o título e os rótulos dos eixos
plt.title('Empréstimos de exemplares por cada ano - 2010 a 2020')
plt.xlabel('Ano do Empréstimo')
plt.ylabel('Quantidade de Empréstimos')

# Exibindo o gráfico
plt.show()
```
