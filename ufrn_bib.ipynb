```python
# Unificando em um dataframe usando pandas, os arquivos de nome emprestimos em formato csv e Apache paquet
# código gerado pelo Gemini  o colab para agilizar o código, que analisei para veificar e entender
# inclui a variável filepath para informar o caminho dos arquivos

import pandas as pd
import glob

# Lista para armazenar os DataFrames de cada arquivo
dfs = []
filepath = '/content/sample_data/ufrn_bib/'
# Loop para ler os arquivos CSV de cada ano
for year in range(2010, 2021):
  for i in range(1, 3):  # Assumindo que você tem dois arquivos para cada ano (emprestimos-2010-1.csv, emprestimos-2010-2.csv, etc.)
    filename = filepath + f'emprestimos-{year}{i}.csv'  # Nome do arquivo
    try:
      # Ler o arquivo CSV e adicionar ao DataFrame
      df = pd.read_csv(filename)
      dfs.append(df)
    except FileNotFoundError:
      print(f"Arquivo {filename} não encontrado.")

# Concatenar todos os DataFrames em um único DataFrame
df_final = pd.concat(dfs, ignore_index=True)

# Imprimir o DataFrame final
print(df_final)
```

```python
# Concatenar o  DataFrame df_final com o arquivo dados_exemplares.parquet em um único Data Frame de nome df_concat, sendo que a coluna codigo_barras é a chave em comum

# Importando em um dataframe df_exemplares 'dados_exemplares.parquet'
df_exemplares = pd.read_parquet(filepath + 'dados_exemplares.parquet')

# Fazendo o Merge dos dois dataframes df_final e df_exemplares em um único dataframe df_concat com a coluna chave sendo codigo_barras
#df_concat = pd.merge(df_final, df_exemplares, on='codigo_barras', how='inner')
df_concat = df_concat.merge(df_exemplares)

# Print do DataFrame
print(df_concat)
```
